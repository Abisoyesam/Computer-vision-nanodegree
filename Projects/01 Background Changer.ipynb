## Background Changer

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('../images/green_screen.jpeg')
img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
img.shape
```

```python
# threshold
lower = np.array([0,170,0])
upper = np.array([0,255,0])

# copy the image
img_copy = np.copy(img)

# create mask
mask = cv2.inRange(img,lower,upper)
plt.imshow(mask,cmap='gray')
```
